```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# gets the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# the names for the columns that will be used
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]

# read the dataset to pandas
iris_df = pd.read_csv(url, header=None, names=columns)

# first 5 rows of the dataset
iris_df.head()
```

```python
# Scatter plots of all the values one vs the other

plt.scatter(iris_df["sepal_length"], iris_df["petal_width"])
plt.xlabel("Sepal Length in cm")
plt.ylabel("Petal Width in cm")
plt.title("Scatter Plot of Sepal Length vs Petal Width")
plt.show()

plt.scatter(iris_df["sepal_length"], iris_df["sepal_width"])
plt.xlabel("Sepal Length in cm")
plt.ylabel("Sepal Width in cm")
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.show()

plt.scatter(iris_df["sepal_length"], iris_df["petal_length"])
plt.xlabel("Sepal Length in cm")
plt.ylabel("Petal Length in cm")
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()

plt.scatter(iris_df["petal_length"], iris_df["petal_width"])
plt.xlabel("Petal Length in cm")
plt.ylabel("Petal Width in cm")
plt.title("Scatter Plot of Petal Length vs Petal Width")
plt.show()

plt.scatter(iris_df["petal_length"], iris_df["sepal_width"])
plt.xlabel("Petal Length in cm")
plt.ylabel("Sepal Width in cm")
plt.title("Scatter Plot of Petal Length vs Sepal Width")
plt.show()

plt.scatter(iris_df["sepal_width"], iris_df["petal_width"])
plt.xlabel("Sepal Width in cm")
plt.ylabel("Petal Width in cm")
plt.title("Scatter Plot of Sepal Width vs Petal Width")
plt.show()


# getting the mean of each value
# and setting it only to 2 decimals with round

sepal_length_mean = iris_df['sepal_length'].mean()
sepal_length_mean_rounded = round(sepal_length_mean, 2)
print("The mean sepal length is",sepal_length_mean_rounded, "cm")

sepal_width_mean = iris_df['sepal_width'].mean()
sepal_width_mean_rounded = round(sepal_width_mean, 2)
print("The mean sepal width is",sepal_width_mean_rounded, "cm")

petal_length_mean = iris_df['petal_length'].mean()
petal_length_mean_rounded = round(petal_length_mean, 2)
print("The mean petal length is",petal_length_mean_rounded, "cm")

petal_width_mean = iris_df['petal_width'].mean()
petal_width_mean_rounded = round(petal_width_mean, 2)
print("The mean petal width is",petal_width_mean_rounded, "cm")

```

```python
# bar charts that shows all the characteristics per flower

plt.bar(iris_df.index, iris_df["sepal_length"], color='black')
plt.xlabel("Flower")
plt.ylabel("Sepal Length")
plt.title("Sepal Lenght per Flower")
plt.show()

plt.bar(iris_df.index, iris_df["petal_length"], color='black')
plt.xlabel("Flower")
plt.ylabel("Petal Length")
plt.title("Petal Lenght per Flower")
plt.show()

plt.bar(iris_df.index, iris_df["sepal_width"], color='black')
plt.xlabel("Flower")
plt.ylabel("Sepal Width")
plt.title("Sepal Width per Flower")
plt.show()

plt.bar(iris_df.index, iris_df["petal_width"], color='black')
plt.xlabel("Flower")
plt.ylabel("Petal Width")
plt.title("Petal Width per Flower")
plt.show()
```
