```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# the names for the columns that will be used
columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]

# read the dataset to pandas
iris_df = pd.read_csv('iris.data', header=None, names=columns)

# first 5 rows of the dataset
iris_df.head()
```

```python
# Scatter plots of all the values one vs the other

plt.scatter(iris_df["sepal_length"], iris_df["petal_width"])
plt.xlabel("Sepal Length in cm")
plt.ylabel("Petal Width in cm")
plt.title("Scatter Plot of Sepal Length vs Petal Width")
plt.show()

plt.scatter(iris_df["sepal_length"], iris_df["sepal_width"])
plt.xlabel("Sepal Length in cm")
plt.ylabel("Sepal Width in cm")
plt.title("Scatter Plot of Sepal Length vs Sepal Width")
plt.show()

plt.scatter(iris_df["sepal_length"], iris_df["petal_length"])
plt.xlabel("Sepal Length in cm")
plt.ylabel("Petal Length in cm")
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()

plt.scatter(iris_df["petal_length"], iris_df["petal_width"])
plt.xlabel("Petal Length in cm")
plt.ylabel("Petal Width in cm")
plt.title("Scatter Plot of Petal Length vs Petal Width")
plt.show()

plt.scatter(iris_df["petal_length"], iris_df["sepal_width"])
plt.xlabel("Petal Length in cm")
plt.ylabel("Sepal Width in cm")
plt.title("Scatter Plot of Petal Length vs Sepal Width")
plt.show()

plt.scatter(iris_df["sepal_width"], iris_df["petal_width"])
plt.xlabel("Sepal Width in cm")
plt.ylabel("Petal Width in cm")
plt.title("Scatter Plot of Sepal Width vs Petal Width")
plt.show()


# getting the mean of each value
# and setting it only to 2 decimals with round

sepal_length_mean = iris_df['sepal_length'].mean()
sepal_length_mean_rounded = round(sepal_length_mean, 2)
print("The mean sepal length is",sepal_length_mean_rounded, "cm")

sepal_width_mean = iris_df['sepal_width'].mean()
sepal_width_mean_rounded = round(sepal_width_mean, 2)
print("The mean sepal width is",sepal_width_mean_rounded, "cm")

petal_length_mean = iris_df['petal_length'].mean()
petal_length_mean_rounded = round(petal_length_mean, 2)
print("The mean petal length is",petal_length_mean_rounded, "cm")

petal_width_mean = iris_df['petal_width'].mean()
petal_width_mean_rounded = round(petal_width_mean, 2)
print("The mean petal width is",petal_width_mean_rounded, "cm")

```

```python
# bar charts that shows all the characteristics per flower

plt.bar(iris_df.index, iris_df["sepal_length"], color='black')
plt.xlabel("Flower")
plt.ylabel("Sepal Length")
plt.title("Sepal Lenght per Flower")
plt.show()

plt.bar(iris_df.index, iris_df["petal_length"], color='black')
plt.xlabel("Flower")
plt.ylabel("Petal Length")
plt.title("Petal Lenght per Flower")
plt.show()

plt.bar(iris_df.index, iris_df["sepal_width"], color='black')
plt.xlabel("Flower")
plt.ylabel("Sepal Width")
plt.title("Sepal Width per Flower")
plt.show()

plt.bar(iris_df.index, iris_df["petal_width"], color='black')
plt.xlabel("Flower")
plt.ylabel("Petal Width")
plt.title("Petal Width per Flower")
plt.show()
```

```python
#calculating the standard deviations, and plotting the deviations for each characteristic

iris_df_sepal_length_divation = iris_df["sepal_length"].std()
s_l_round = round(iris_df_sepal_length_divation, 2)
print("The standard deviation of the sepal lengths is equal to", s_l_round, "cm.")

sepal_length_deviations = iris_df["sepal_length"] - sepal_length_mean
plt.scatter(iris_df.index, sepal_length_deviations, color='blue', marker='o')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)  # horizontal line at y=0 for reference
plt.title('Deviations from Mean - Sepal Length')
plt.xlabel('Data Point Index')
plt.ylabel('Deviation from Mean')
plt.show()




iris_df_sepal_width_divation = iris_df["sepal_width"].std()
s_w_round = round(iris_df_sepal_width_divation, 2)
print("The standard deviation of the sepal widths is equal to", s_w_round, "cm.")

sepal_width_deviations = iris_df["sepal_width"] - sepal_width_mean
plt.scatter(iris_df.index, sepal_width_deviations, color='blue', marker='o')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)  
plt.title('Deviations from Mean - Sepal Width')
plt.xlabel('Data Point Index')
plt.ylabel('Deviation from Mean')
plt.show()

iris_df_petal_length_divation = iris_df["petal_length"].std()
p_l_round = round(iris_df_petal_length_divation, 2)
print("The standard deviation of the petal lengths is equal to", p_l_round, "cm.")

petal_length_deviations = iris_df["petal_length"] - petal_length_mean
plt.scatter(iris_df.index, petal_length_deviations, color='blue', marker='o')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)  
plt.title('Deviations from Mean - Sepal Width')
plt.xlabel('Data Point Index')
plt.ylabel('Deviation from Mean')
plt.show()

iris_df_petal_width_divation = iris_df["petal_width"].std()
p_w_round = round(iris_df_petal_width_divation, 2)
print("The standard deviation of the petal widths is equal to", p_w_round, "cm.")

petal_width_deviations = iris_df["petal_width"] - petal_width_mean
plt.scatter(iris_df.index, petal_width_deviations, color='blue', marker='o')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)  
plt.title('Deviations from Mean - Sepal Width')
plt.xlabel('Data Point Index')
plt.ylabel('Deviation from Mean')
plt.show()

```

```python
# creates a histogram for the distribution of each characteristic
sepal_lengths = iris_df["sepal_length"]

sns.histplot(sepal_lengths, kde=True)  # adds a smooth line across the plot
plt.title('Distribution of Sepal Lengths')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()


sepal_widths = iris_df["sepal_width"]

sns.histplot(sepal_widths, kde=True)  
plt.title('Distribution of Sepal Widths')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()


petal_lengths = iris_df["petal_length"]

sns.histplot(petal_lengths, kde=True) 
plt.title('Distribution of Petal Lengths')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()


petal_widths = iris_df["petal_width"]

sns.histplot(petal_widths, kde=True)  
plt.title('Distribution of Petal Widths')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```
